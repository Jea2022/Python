# Data Cleaning in Pandas

```python
#import library
import pandas as pd
```

```python
df = pd.read_excel(r'C:\Users\ASUS\OneDrive\Desktop\python project\Customer Call List.xlsx')
```

```python
df
```

```python
#remove duplicate value
df = df.drop_duplicates()
df
```

```python
df = df.drop(columns = 'Not_Useful_Column')
df
```

```python
#strip, lstrip, rstrip ==> remove extra white space and other character from the left and right of the column
df['Last_Name'] = df['Last_Name'].str.lstrip("/")
df['Last_Name'] = df['Last_Name'].str.lstrip("...") #remove white space from the left sied
df['Last_Name'] = df['Last_Name'].str.rstrip("_") #remove white space from the right side
df

#the easy way
df['Last_Name'] = df['Last_Name'].str.rstrip("123._/") #this methods removes all of them which is inside the qoutes
```

```python
#Standardize the phone number into specific format
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]','')
df
```

```python
#standardize the phone number using lamda expression
df['Phone_Number'] = df['Phone_Number'].apply(lambda x:str(x)) #convert every value into string data type
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10]) #split every 3 digits with -
df
```

```python
#replace the value
df['Phone_Number'] = df['Phone_Number'].str.replace('nan----', '')
df['Phone_Number'] = df['Phone_Number'].str.replace('Na----', '')
df
```

```python
#split the column into individual base on the dilimeter
df['Address'].str.split(',', 2 , expand = True)
```

```python
#name every individual columns
df[['Stree Address','State','Zip']] = df['Address'].str.split(',', 2 , expand = True)
```

```python
df
```

```python
#Replace the sepcific value 
df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N')
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y')

df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No', 'N')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes', 'Y')
df
```

```python
df = df.replace('N/a','')
df
```

```python
#fill the empty column
df = df.fillna('')
df
```

```python
#remove the when the do-not-contact is yes (Y)
for x in df.index:
    if df.loc[x, 'Do_Not_Contact'] == 'Y':
        df.drop(x, inplace = True)
        
df
```

```python
#remove the phone number when this is empty
for x in df.index:
    if df.loc[x, 'Phone_Number'] == '':
        df.drop(x, inplace = True)
        
df
```

```python
#drop the 'Address' column
df = df.drop(columns = 'Address')
df
```
